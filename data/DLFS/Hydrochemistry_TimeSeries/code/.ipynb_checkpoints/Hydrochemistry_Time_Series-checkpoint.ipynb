```python
# url to data stored on github repository
url = "https://raw.githubusercontent.com/WFB2/BoknisEckTS/main/BoknisEck_1957-2014.tsv" 
 
# open dataframe with pandas
hydrochem_data = pd.read_csv(url, sep='\t', header=(0))   
```

## Data preparation

Preparing the dataset by dropping unwanted columns and rows with NAs in them and one_hot encoding


### Delete NAs and not needed Variables

```python
 
# delete flag, long, lat, Cast columns (axis = 1) as not needed
 
hydrochem_data = hydrochem_data.drop(["Flag (SiO2)", "Flag (NO3)", "Flag (NO2)", "Flag (Oxygen)", "Flag (PO4)", "Flag (SiO2)", "Latitude", "Longitude", "Cast"], axis = 1 )
 
 
# drop rows with NAs
 
hydrochem_data = hydrochem_data.dropna()






```

### Prepare Time Variables (Week, Month, Daylength)

```python
# converting date to week number and one_hot encoding it

hydrochem_data['Date/Time'] =pd.to_datetime(hydrochem_data['Date/Time'], errors ='coerce') # Change date format, eliminating time, needed for next step (MB)
hm_date = hydrochem_data.copy()  # Saving dataframe with date for heat map (JB)
hydrochem_data['week'] =hydrochem_data['Date/Time'].dt.week  # changing date format to week number (MB)
hydrochem_data['month'] =hydrochem_data['Date/Time'].dt.month

date_time = pd.to_datetime(hydrochem_data['Date/Time'], errors ='coerce')


week_onehot = pd.get_dummies(hydrochem_data.week, prefix= "week") # one_hot encoding the week number, week is now called week_onehot, is a separate dataframe (MB)
# 2 dataframes from here, hydrochem_data and week__onehot, put back together after normalization. (MB)

month_onehot = pd.get_dummies(hydrochem_data.month, prefix= "month")
```

```python
## Add Daylength

hydrochem_data['Daylength'] = 0
t = pd.DatetimeIndex(hydrochem_data['Date/Time'])

for i in range(0, hydrochem_data.shape[0]):
  hydrochem_data['Daylength'].iloc[i] = location.durationtuple(datetime(t[i].year, t[i].month, t[i].day))[0]*60 + location.durationtuple(datetime(t[i].year, t[i].month, t[i].day))[1] 
```

```python
hydrochem_data = hydrochem_data.drop(['Date/Time', 'week'], axis = 1 ) # dropping unneeded date columns (MB)
```

### One-Hot Encode Depth and Chl_a

```python
## Group the depth

hydrochem_data['Depth water [m]'] 
hydrochem_data['Depth_grouped'] = hydrochem_data['Depth water [m]'].apply(lambda x: 1 if x <= 3
                                                                else (5 if (x > 3 and x <= 7)
                                                                else (10 if (x > 7 and x <= 12)
                                                                else (15 if (x > 12 and x <= 17)
                                                                else (20 if (x > 17 and x <= 22)
                                                                else 25 )))))
                                                                
hydrochem_data['Depth_grouped'].unique()                                                             
Depth_onehot = pd.get_dummies(hydrochem_data.Depth_grouped, prefix= "Depth")

#hydrochem_data = hydrochem_data.drop(['Depth water [m]'], axis = 1)
```

```python
Depth_onehot.head()
```

```python
# divide chl_a into low medium and high by using the 33 and 66 quantiles
chl_1 = hydrochem_data['Chl a [Âµg/l]'].quantile(0.2) 
chl_2 = hydrochem_data['Chl a [Âµg/l]'].quantile(0.4) 
chl_3 = hydrochem_data['Chl a [Âµg/l]'].quantile(0.6) 
chl_4 = hydrochem_data['Chl a [Âµg/l]'].quantile(0.8)  



# substituting the values with low, medium and high classifications

hydrochem_data['Chl_classified'] = hydrochem_data['Chl a [Âµg/l]'].apply(lambda x: 'very_low' if x <= chl_1
                                                                else ('low' if (x > chl_1 and x <= chl_2)
                                                                else ('medium' if (x > chl_2 and x <= chl_3)
                                                                else ('high' if (x > chl_3 and x <= chl_4)
                                                                else 'very_high' ))))


# putting the chl_classified into own dataframe and removing it from hydrochem_data, as it results in NAs after normalisation

chl_classified = hydrochem_data['Chl_classified']
hydrochem_data = hydrochem_data.drop(['Chl_classified'], axis = 1)

```

```python
chl_classified.head(5)
```

### First model: Mean normalisation

#### Assemble the variables for the model

```python
# New Dataframe for Model 1
model_1_old_data = hydrochem_data_norm_old
```

```python
# Add One-Hot Encoded Dataframes

#model_1_old_data = pd.concat([model_1_old_data, Depth_onehot], axis = 1)
#model_1_old_data = pd.concat([model_1_old_data, week_onehot], axis = 1) 
model_1_old_data = pd.concat([model_1_old_data, month_onehot], axis = 1)
```

```python
model_1_old_data.head(5)
```

```python
# Drop variables from Dataset:


#model_1_old_data = model_1_old_data.drop(columns = "Depth_grouped")
model_1_old_data = model_1_old_data.drop(columns = "Depth water [m]")
#model_1_old_data = model_1_old_data.drop(columns = "Chl_classified")
model_1_old_data = model_1_old_data.drop(columns = "Sample label")
#model_1_old_data = model_1_old_data.drop(columns = "[NO2]- [Âµmol/l]")
#model_1_old_data = model_1_old_data.drop(columns = "[NO3]- [Âµmol/l]")
#model_1_old_data = model_1_old_data.drop(columns = "O2 [Âµmol/kg]")
#model_1_old_data = model_1_old_data.drop(columns = "[PO4]3- [Âµmol/l]")
model_1_old_data = model_1_old_data.drop(columns = "Sal")
#model_1_old_data = model_1_old_data.drop(columns = "SiO2 [Âµmol/l]")
#model_1_old_data = model_1_old_data.drop(columns = "Temp [Â°C]")
#model_1_old_data = model_1_old_data.drop(columns = "Daylength")




model_1_old_data.head()
```

#### Splitting the data into training and test set

```python
# shuffle data and separate into data and labels+

shuffled_hydrochem_data = shuffle(model_1_old_data)

Y = shuffled_hydrochem_data.pop("Chl a [Âµg/l]").to_frame()
X = shuffled_hydrochem_data

# split into train and test
X_train_old, X_test_old, Y_train_old, Y_test_old = train_test_split(X, Y, test_size=0.2)

X_train_old_month = X_train_old.pop("month").to_frame()
X_test_old_month = X_test_old.pop("month").to_frame()


print ("There are", X_train_old.shape[0], "datapoints in the training set. \n There are", X_test_old.shape[0], "datapoints in the test set." )

X_train_old.head(10)
```

#### Setting up the model

```python
# function that sets up the model
def create_model(input_shape):
    #tf.keras.backend.set_floatx('float64')
    # creation of the model
    model = tf.keras.models.Sequential([
        
        tf.keras.layers.Input(shape=input_shape),
        #normalizer,
        tf.keras.layers.Dense(64, activation="relu"),
        #tf.keras.layers.Dropout(0.2),
      
       tf.keras.layers.Dense(16, activation="relu"),
       #tf.keras.layers.Dropout(0.2),
        
       
        

        # last layer needs to have a linear activation function because we solve a regression problem
        tf.keras.layers.Dense(1, activation="linear")
    ])
    # compile model
    opt = tf.keras.optimizers.Adam(learning_rate = 0.0001)
    model.compile(optimizer= opt, loss='mean_absolute_error', metrics=["mean_squared_error"])
    return model
```

#### Training the model

```python
model_1_old = create_model(X_train_old.shape[1])
history_1_old = model_1_old.fit(X_train_old, Y_train_old, epochs=1000, validation_split=0.1, verbose = 1)
```

#### Evaluating the trained model

```python
def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.ylim([0, 0.8])
  plt.xlabel('Epoch')
  plt.ylabel('Loss')
  plt.legend()
  plt.grid(True)
```

One simple rule to remember: use categorical crossentropy loss when your Neural Network dataset has one-hot encoded target values!

```python
# split into train and test
X_2_train, X_2_test, Y_2_train, Y_2_test = train_test_split(X, Y, test_size=0.2)


# one_hot encoding chl 
Y_2_train = pd.get_dummies(Y_2_train, prefix = 'chl')
Y_2_test = pd.get_dummies(Y_2_test, prefix = 'chl')
```

```python
#X_train.head(10)
#X_train.describe().transpose()
Y_2_train.shape
```

#### Setting up the model

```python
# function that sets up the model
def create_model(input_shape):
    #tf.keras.backend.set_floatx('float64')
    # creation of the model
    model = tf.keras.models.Sequential([
        
        tf.keras.layers.Input(shape=input_shape),
        #normalizer,
        tf.keras.layers.Dense(32, activation="relu"),
        tf.keras.layers.Dropout(0.4),
      
       tf.keras.layers.Dense(16, activation="relu"),
       tf.keras.layers.Dropout(0.4),
        

        
        #last layer softmax as we have 3 variables...?
        tf.keras.layers.Dense(5, activation="softmax")

        # last layer needs to have a linear activation function because we solve a regression problem
        #tf.keras.layers.Dense(1, activation="linear")
    ])
    # compile model
    opt = tf.keras.optimizers.Adam(learning_rate = 0.001)

    model.compile(optimizer= opt, loss='categorical_crossentropy', metrics=["accuracy"])
    return model
```

#### Training the model

```python
model_2 = create_model(X_2_train.shape[1])
history_2 = model_2.fit(X_2_train, Y_2_train, epochs=500, validation_split=0.2, verbose = 1)
```

```python
model_3_data.head(5)
```

```python
# Drop variables from Dataset:


#model_3_data = model_3_data.drop(columns = "Depth_grouped")
model_3_data = model_3_data.drop(columns = "Depth water [m]")
model_3_data = model_3_data.drop(columns = "Sample label")
#model_3_data = model_3_data.drop(columns = "[NO2]- [Âµmol/l]")
model_3_data = model_3_data.drop(columns = "[NO3]- [Âµmol/l]")
#model_3_data = model_3_data.drop(columns = "O2 [Âµmol/kg]")
model_3_data = model_3_data.drop(columns = "[PO4]3- [Âµmol/l]")
model_3_data = model_3_data.drop(columns = "Sal")
model_3_data = model_3_data.drop(columns = "SiO2 [Âµmol/l]")
#model_3_data = model_3_data.drop(columns = "Temp [Â°C]")
model_3_data = model_3_data.drop(columns = "Daylength")



model_3_data.head()
```

#### Splitting the data into training and test set

```python
# shuffle data and separate into data and labels+

shuffled_hydrochem_data = shuffle(model_3_data)

Y = shuffled_hydrochem_data.pop("Chl a [Âµg/l]").to_frame()
X = shuffled_hydrochem_data

# split into train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

X_train_month = X_train.pop("month").to_frame()
X_test_month = X_test.pop("month").to_frame()

print ("There are", X_train.shape[0], "datapoints in the training set. \n There are", X_test.shape[0], "datapoints in the test set." )

X_train.head(10)
```

#### Setting up the model

```python
# function that sets up the model
def create_model(input_shape):
    
    model = tf.keras.models.Sequential([
        
        tf.keras.layers.Input(shape=input_shape),
        
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.Dropout(0.2),
      
        tf.keras.layers.Dense(8, activation="relu"),
        tf.keras.layers.Dropout(0.2),
        
        tf.keras.layers.Dense(1, activation="linear")
    ])
    # compile model
    opt = tf.keras.optimizers.Adam(learning_rate = 0.0001)
    model.compile(optimizer= opt, loss='mean_absolute_error', metrics=["mean_squared_error"])
    return model
```

#### Training the model

```python
model_3 = create_model(X_train.shape[1])
history_3 = model_3.fit(X_train, Y_train, epochs=1000, validation_split=0.1, verbose = 1)
```
